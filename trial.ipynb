```python
new_st_data = st_data["review_year"]==2013
grouped_sum = st_data.groupby("review_year")[["number_of_reviews"]].sum()
grouped_sum.reset_index()
grouped_sum_chart = px.line(grouped_sum,x=year_list,y="number_of_reviews",title="ABC")
grouped_sum_chart.show()
```
